```python
all_items = set(train_df.category.values)
all_items_dict = {k:v for v,k in enumerate(all_items)}
all_items_dict
```

```python
len(train_df.user_id.unique())
```

```python
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
aov_data = train_df.aov.to_numpy()
scalar.fit(aov_data.reshape(-1,1))
train_df.aov = scalar.transform(train_df.aov.to_numpy().reshape(-1,1))
train_df.head()
```

Train_df has 29972 user data but train_target_df has only 13245 user data

```python
new_df = pd.DataFrame(data=train_target_df.user_id, columns=[['user_id']])
for items in all_items:
  new_df[items] = np.nan

new_df
```

```python
row_len = new_df.shape
```

```python
for i,row in enumerate(train_target_df.iterrows()):
  print(type(row))
  row_val = np.zeros(19)
  row_val[0] = row[1].user_id
  temp_df = train_df[train_df.user_id == row_val[0]]
  print(temp_df)
  for temp_df_row in temp_df.iterrows():
    category = temp_df_row[1].category
    row_val[all_items_dict[category] + 1] += temp_df_row[1].aov #all_item_dict is 0 indexed but here row[0] = user-id

  new_df.loc[i] = row_val

new_df.head()
```

```python
new_df.shape
```

```python
#label encoding, let's see
new_labels = train_target_df.category
print(type(new_labels))
new_labels = new_labels.map(all_items_dict)
new_labels
```

```python
X, y = new_df.drop(columns=['user_id'], axis=1), new_labels
from xgboost import XGBClassifier
clf = XGBClassifier()
clf.fit(X,y)
```

```python
clf.score(X,y)
```

```python
sam_sub = pd.read_csv('/content/Sample Submission.csv')
sam_sub.head()
```

```python
pred = clf.predict_proba(new_test_df.drop(columns=['user_id'], axis=1))
```

```python
pred_top_3 = []
for pre in pred:
  temp_arr = []
  pre = pre.tolist()
  for i in range(3):
    index = pre.index(max(pre))
    pre[index] = -1
    temp_arr.append(index)

  pred_top_3.append(temp_arr)
```

```python
pred_top_3_str = []
inverse_dict = {v:k for k,v in all_items_dict.items()}
for pre in pred_top_3:
  s = inverse_dict[pre[0]] + ', ' + inverse_dict[pre[1]] + ', ' + inverse_dict[pre[2]]
  pred_top_3_str.append(s)

#pred_top_3_str
```
